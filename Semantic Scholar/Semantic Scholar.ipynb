```python

# Packages
from bs4 import BeautifulSoup
import requests
import pandas as pd
import regex as re



# Request to Semanticscholar
url = 'https://semanticscholar.org/search?q=Computational%20Propaganda&sort=relevance'
page = requests.get(url)
soup1 = BeautifulSoup(page.content, 'html.parser')
# soup1


# # Open html from a file
with open("/Users/tomasnagy/Documents/Weizenbaum - Work/Semantic Scholar/computational propaganda _ Semantic Scholar.html") as file:
    soup = BeautifulSoup(file, 'html.parser')
# soup
```

```python
# Get the Title
results = []
for entry in soup.find_all('h2', class_='cl-paper-title'):
    results.append({"title": entry.text})  
dataframe1 = pd.DataFrame(results)
```

```python
# Get the Authors
results2 = []
for entry in soup.find_all('span', class_='cl-paper-authors'):
    results2.append({"author": entry.text})    
dataframe2 = pd.DataFrame(results2)
```

```python
# Get the Publication Date
results3 = []
for entry in soup.find_all('span', class_='cl-paper-pubdates'):
    results3.append({"publication date": entry.text})    
dataframe3 = pd.DataFrame(results3)
```

```python
results3 = []
for entry in soup.find_all('span', class_='cl-paper-pubdates'):
    results3.append({"publication dation": entry.text})    
dataframe3 = pd.DataFrame(results3)
```

```python
# Get the Citations
results4 = []
for entry in soup.find_all('div', class_='cl-paper__bulleted-row cl-paper-controls'):
    if entry.find_all("div", {"class": "cl-paper-stats__action-stat"}):
        for item in entry.find_all("div", {"class": "cl-paper-stats__action-stat"}):
            results4.append({"citation": item.text})   
    else:
        results4.append({"citation": '0'}) 
dataframe4 = pd.DataFrame(results4)
```

```python
# Get the Publisher
results5 = []
for entry in soup.find_all('ul', class_='cl-paper__bulleted-row'):
    if entry.find_all("span", {"class": "cl-paper-fos"}):
        for item in entry.find_all("span", {"class": "cl-paper-fos"}):
            results5.append({"publisher": item.text})   
    else:
        results5.append({"publisher": 'No publisher found'})
dataframe5 = pd.DataFrame(results5)
```

```python
# Get the Link
results6 = []

for entry in soup.find_all('div', class_ = 'flex-row cl-paper-actions'):
    if entry.find_all('a', href=True):
        for a in entry.find_all('a', href=True):
            results6.append({"link": a['href']})
    else:
        results6.append({"link": 'Link not found'})
dataframe6 = pd.DataFrame(results6)
```

```python
# Merge everything into one dataframe
dataframe = pd.concat([dataframe1, dataframe2,dataframe3,dataframe4,dataframe5,dataframe6], axis=1)
dataframe.to_csv('Smenatic Scholar - publications', sep='\t')
dataframe
```
